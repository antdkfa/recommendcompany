```python
import requests
from bs4 import BeautifulSoup

real_data = []
company = "네이버"
for i in range(2,10):
    univ_data = []
    tmp_data = []
    score_data = []
    j = 0
    response = requests.get(f'http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=924&VPage={i}')
    soup = BeautifulSoup(response.content, 'html.parser')
    div_indLists = soup.find('div',{'class' : 'indLists'})
    univ_data = div_indLists.find_all('span', {'class' : 'tx'})
    div_totalList = div_indLists.find_all('div', {'class' : 'totalList'})
    for spec_data in div_totalList:
        score_data.append(list(spec_data.find_all('span', {'class' : 'score'})))
    for txt in score_data:
        univ = univ_data[j].text
        if univ[0] == '해':
            univ = "해외"
        if univ[0] == '수':
            univ = "수도권"
        if univ[0] == '서':
            univ = "서울"
        if univ[0] == '지':
            univ = "지방"
        grade = txt[0].text
        if grade[2] == '-':
            grade = "00.00"
        toeic = txt[1].text
        if toeic == '-':
            toeic = '0'
        toss = txt[2].text
        if toss == '-':
            toss = '0' 
        opic = txt[3].text
        if opic == '-':
            opic = '0' 
        other_lang = txt[4].text
        if other_lang == '-':
            other_lang = '0' 
        cert = txt[5].text
        if cert == '-':
            cert = '0' 
        oversea_exp = txt[6].text
        if oversea_exp == '-': oversea_exp = '0' 
        intern = txt[7].text
        if intern == '-':
            intern = '0' 
        award = txt[8].text
        if award == '-': 
            award = '0' 
        vol = txt[9].text
        if vol == '-':
            vol = '0' 
        tmp_data.append([univ, grade[1] + grade[2] + grade[3], toeic, toss, opic, other_lang, cert, oversea_exp, intern, award, vol, company])
        j += 1
        real_data.extend(tmp_data)
        tmp_data = []

for real in real_data:
    print(real)
```
