```python
import requests
from bs4 import BeautifulSoup

real_data = []
company = [[] * 15 for i in range(30)]
link = [[] * 15 for i in range(100)]
company[0] = "카카오"
company[1] = "네이버"
company[2] = "삼성전자"
company[3] = "삼성SDS"
company[4] = "SK하이닉스"
company[5] = "kt"
company[6] = "lg유플러스"
company[7] = "롯데정보통신"
company[8] = "CJ올리브네트웍스"
company[9] = "NC소프트"
link[0] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=924&VPage=" 
link[1] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=215&VPage="
link[2] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=1&VPage="
link[3] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=129&VPage="
link[4] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=42&VPage="
link[5] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=23&VPage="
link[6] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=1628&VPage="
link[7] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=677&VPage="
link[8] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=2188&VPage="
link[9] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=513&VPage="
link[10] = "http://www.jobkorea.co.kr/starter/spec/view?Page=1&IsAlumniPersonOn=0&schEdu=50&Tab=3&C_Idx=28&VPage="
for k in range(0,10):
    for i in range(2,16):
        univ_data = []
        tmp_data = []
        score_data = []
        j = 0
        response = requests.get(f'{link[k]}{i}')
        soup = BeautifulSoup(response.content, 'html.parser')
        div_indLists = soup.find('div',{'class' : 'indLists'})
        univ_data = div_indLists.find_all('span', {'class' : 'tx'})
        div_totalList = div_indLists.find_all('div', {'class' : 'totalList'})
        for spec_data in div_totalList:
            score_data.append(list(spec_data.find_all('span', {'class' : 'score'})))
        for txt in score_data:
            univ = univ_data[j].text
            if univ[0] == '해':
                univ = "해외"
            if univ[0] == '수':
                univ = "수도권"
            if univ[0] == '서':
                univ = "서울"
            if univ[0] == '지':
                univ = "지방"
            grade = txt[0].text
            if grade[2] == '-':
                grade = "00.00"
            toeic = txt[1].text
            if toeic == '-':
                toeic = '0'
            toss = txt[2].text
            if toss == '-':
                toss = '0' 
            opic = txt[3].text
            if opic == '-':
                opic = '0' 
            other_lang = txt[4].text
            if other_lang == '-':
                other_lang = '0' 
            cert = txt[5].text
            if cert == '-':
                cert = '0' 
            oversea_exp = txt[6].text
            if oversea_exp == '-': oversea_exp = '0' 
            intern = txt[7].text
            if intern == '-':
                intern = '0' 
            award = txt[8].text
            if award == '-': 
                award = '0' 
            vol = txt[9].text
            if vol == '-':
                vol = '0' 
            tmp_data.append([univ, grade[1] + grade[2] + grade[3], toeic, toss, opic, other_lang, cert, oversea_exp, intern, award, vol, company[k]])
            j += 1
        real_data.extend(tmp_data)
        
print(real_data)
```

```python
from pandas import DataFrame

df = DataFrame(real_data, columns= ['학교명', '학점', '토익', '토스', '오픽','외국어', '자격증', '수상경험', '인턴', '해외경험', '봉사활동', '기업명'])
export_excel = df.to_csv(r'C:\Users\antdk\Desktop\data.csv') 
```
